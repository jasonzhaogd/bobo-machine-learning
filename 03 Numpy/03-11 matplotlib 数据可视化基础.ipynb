## 折线图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
z = np.cos(x)
```

```python
plt.plot(x, y)
plt.plot(x, z)
```

```python
plt.plot(x, y)
plt.plot(x, z, color="red", linestyle="--")
```

```python
plt.plot(x, y)
plt.plot(x, z, color="red", linestyle="--")
plt.xlim(-5, 15)
plt.ylim(-3, 3)
```

```python
plt.plot(x, y)
plt.plot(x, z, color="red", linestyle="--")
plt.axis([-5, 15, -3, 3])
```

```python
plt.plot(x, y, label="sin(x)")
plt.plot(x, z, label="cox(x)", color="red", linestyle="--")
plt.legend()
plt.xlabel("X 轴")
plt.ylabel("Y 轴")
```

```python
plt.title("matplotlib 基础")
plt.plot(x, y, label="sin(x)")
plt.plot(x, z, label="cox(x)", color="red", linestyle="--")
plt.legend()
plt.xlabel("X 轴")
plt.ylabel("Y 轴")
```
